```python
import transformers
import numpy as np
import torch
```

```python
name = 'facebook/bart-base'
generator = transformers.BartForConditionalGeneration.from_pretrained(name)
tokenizer = transformers.BartTokenizer.from_pretrained(name)
```

```python
# prompt
prompt = "The dog ate a <mask> and <mask> it greatly"
input_ids = tokenizer.encode(prompt, return_tensors='pt')
```

```python
# embeddings
embed_matrix = generator.get_input_embeddings()
embeddings = embed_matrix(input_ids)
```

```python
# chooses input_embeds
inputs_embeds = embeddings
```

```python
# defaults
attention_mask = torch.ones(inputs_embeds.shape[:2], dtype=torch.long)
decoder_input_ids = torch.ones((inputs_embeds.shape[0], 1), dtype=torch.long)*generator.config.decoder_start_token_id
```

```python
# generates output?!
output = generator(attention_mask=attention_mask, decoder_input_ids=decoder_input_ids, inputs_embeds=inputs_embeds)
```

```python
# samples next 10 tokens from output logits
logits = output.logits
next_token_logits = logits[:, -1, :]
next_token = torch.argmax(next_token_logits, dim=-1)
```
